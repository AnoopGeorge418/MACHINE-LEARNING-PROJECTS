# **Data Collection and Preprocessing**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn.datasets
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
house = pd.read_csv("/content/housing.csv", header = None, delimiter = "\s+")
house.head()
```

```python
house.shape
```

```python
house.info()
```

```python
house.describe()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(house.corr(), annot = True, cbar = True, square = True, fmt= '.1f', annot_kws={'size':8}, cmap = 'Blues')
```

```python
# separating
X = house.iloc[:,:-1]
Y = house.iloc[:, -1]
```

```python
# splitting data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
# Model prediction
y_pred = model.predict(X_train)
y_pred
```

```python
# Evaluation model - r2
score_1 = metrics.r2_score(y_train, y_pred)
print("R2 score : ", score_1)

score_2 = metrics.mean_absolute_error(y_train, y_pred)
print("Mean Absolute Error : ", score_2)
```

```python
y_pred_test = model.predict(X_test)
y_pred_test
```

```python
# Evaluation model - r2
score_1 = metrics.r2_score(y_test, y_pred_test)
print("R2 score : ", score_1)

score_2 = metrics.mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error : ", score_2)
```
